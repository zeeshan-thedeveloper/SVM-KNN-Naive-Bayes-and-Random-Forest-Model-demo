```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from PIL import Image

# Create a folder for storing results
algo_name = "SVM"
results_dir = "Results"
algo_dir = os.path.join(results_dir, algo_name)

# Check if the directory exists, if not, create it
if not os.path.exists(algo_dir):
    os.makedirs(algo_dir)

# Data for the 10 folds of SVM-Linear Kernel
folds = list(range(1, 11))
accuracy_svm = [91.90, 91.20, 91.30, 90.00, 92.20, 91.70, 90.90, 91.30, 91.60, 91.70]
time_taken_svm = [25.92, 27.60, 22.72, 26.62, 27.06, 29.50, 27.31, 25.37, 19.94, 20.45]

# Calculate summary statistics for SVM-Linear Kernel
mean_accuracy_svm = np.mean(accuracy_svm)
accuracy_variance_svm = np.var(accuracy_svm)
total_time_svm = np.sum(time_taken_svm)
best_accuracy_svm = np.max(accuracy_svm)

# Create a DataFrame for SVM results
data_svm = {
    "Fold": folds,
    "Accuracy (%)": accuracy_svm,
    "Time Taken (s)": time_taken_svm
}
df_svm = pd.DataFrame(data_svm)

# Plotting table for SVM-Linear Kernel results
plt.figure(figsize=(8, 4))
plt.axis('tight')
plt.axis('off')
table_svm = plt.table(cellText=df_svm.values, colLabels=df_svm.columns, loc='center', cellLoc='center')
table_svm.auto_set_font_size(False)
table_svm.set_fontsize(10)
table_svm.auto_set_column_width(col=list(range(len(df_svm.columns))))
plt.title("SVM-Linear Kernel: 10-Fold Cross-Validation Summary", fontsize=14, fontweight='bold')

# Save the table image in the Results/SVM directory
table_image_path = os.path.join(algo_dir, "svm_fold_summary_table.png")
plt.savefig(table_image_path)
plt.close()

# Plotting graph for accuracy variance for SVM-Linear Kernel
plt.figure(figsize=(10, 6))
plt.plot(folds, accuracy_svm, marker='o', linestyle='-', color='blue', label='Accuracy (%)')
plt.axhline(mean_accuracy_svm, color='red', linestyle='--', label=f'Mean Accuracy ({mean_accuracy_svm:.2f}%)')
plt.fill_between(folds, mean_accuracy_svm - np.sqrt(accuracy_variance_svm), 
                 mean_accuracy_svm + np.sqrt(accuracy_variance_svm), 
                 color='blue', alpha=0.2, label='Variance Range')

plt.title("SVM-Linear Kernel: Accuracy Variance Across Folds", fontsize=14, fontweight='bold')
plt.xlabel("Fold Number", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(folds)
plt.tight_layout()

# Save the variance graph in the Results/SVM directory
graph_image_path = os.path.join(algo_dir, "svm_accuracy_variance_graph.png")
plt.savefig(graph_image_path)
plt.close()

# Print summary for SVM-Linear Kernel results
print("\n--- SVM-Linear Kernel Results ---")
print(f"Mean Accuracy: {mean_accuracy_svm:.4f}%")
print(f"Accuracy Variance: {accuracy_variance_svm:.4f}")
print(f"Total Time for All Folds: {total_time_svm:.2f} seconds")
print(f"Best Accuracy: {best_accuracy_svm:.4f}%")

# Output the paths to the saved files
print(f"Results saved in: {algo_dir}")
print(f"Table image: {table_image_path}")
print(f"Variance graph image: {graph_image_path}")

# Show the results (images)
print("\nDisplaying results...")

# Open and display the saved table image
table_img = Image.open(table_image_path)
table_img.show()

# Open and display the saved variance graph image
graph_img = Image.open(graph_image_path)
graph_img.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the algorithms
algorithms = ['SVM Linear', 'SVM Polynomial', 'SVM RBF', 'Random Forest', 'Naive Bayes', 'k-NN Euclidean', 'k-NN Manhattan', 'k-NN Cosine']
accuracies = [91.38, 94.30, 93.50, 94.83, 34.68, 91.23, 92.98, 91.34]  # Best accuracy values for each algorithm
times = [25.25, 29.87, 30.81, 27.81, 2.33, 418.52, 225.07, 631.69]  # Best time values for each algorithm

# Bar width
bar_width = 0.35
index = np.arange(len(algorithms))

# Create a figure and axis for plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Accuracy Bar Chart
ax1.bar(index, accuracies, bar_width, color='skyblue')
ax1.set_title('Accuracy Comparison for MNIST (Best Results)')
ax1.set_xlabel('Algorithms')
ax1.set_ylabel('Accuracy (%)')
ax1.set_xticks(index)
ax1.set_xticklabels(algorithms, rotation=45)
ax1.set_ylim(30, 100)

# Time Bar Chart
ax2.bar(index, times, bar_width, color='lightcoral')
ax2.set_title('Time Comparison for MNIST (Best Results)')
ax2.set_xlabel('Algorithms')
ax2.set_ylabel('Time (seconds)')
ax2.set_xticks(index)
ax2.set_xticklabels(algorithms, rotation=45)
ax2.set_ylim(0, max(times) + 50)

# Save the plot as an image
plt.tight_layout()
plt.savefig('mnist_algorithm_comparison.jpg')  # Save as a .jpg file

# Display the plot
plt.show()
```
